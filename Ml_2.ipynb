```
left=pd.DataFrame({'key':['k0','k1','k2','k4','k5'],
                   'A':['A0','A1','A2','A3','A4','A5'],
                   'B':['B0','B1','B2','B3','B4','B5']})
```

```
right=pd.DataFrame({'key':['k0','k1','k2','k3','k4','k5'],
                   'C':['C0','C1','C2','C3','C4','C5'],
                   'D':['D0','D1','D2','D3','D4','D5']})
```

```
left.merge(right,on='key')
```

```
left.merge(right,on='key',how='left')
```

```
left=pd.DataFrame({'index':['k0','k1','k2','k3','k4','k5'],
                   'A':['A0','A1','A2','A3','A4','A5'],
                   'B':['B0','B1','B2','B3','B4','B5']})

right=pd.DataFrame({'index':['k0','k1','k2','k3','k4','k5'],
                   'C':['C0','C1','C2','C3','C4','C5'],
                   'D':['D0','D1','D2','D3','D4','D5']})
```

```
left.join(right)
```
